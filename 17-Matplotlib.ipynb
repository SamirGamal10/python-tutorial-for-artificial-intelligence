```python
#  sin
import numpy as np 
plt.style.use ('ggplot')
x = np.linspace(0, 10) 
y = np.sin(x)
plt.plot(x, y)
```

```python
# cos
import numpy as np 
plt.style.use ('ggplot')
x = np.linspace(0, 10) 
y = np.cos(x) 
plt.plot(x, y)
```

```python
# xlabel and ylabel
a = (1,2,3,4,5,6)
b = (3,9,10,12,16,18) 
plt.style.use ('ggplot') 
plt.ylabel('Time') 
plt.xlabel('Work')
plt.plot(a,b)
```

```python
# set_xlabel and set_ylabel
a = (1,2,3,4,5,6)
b = (3,9,10,12,16,18) 
plt.style.use ('ggplot') 
ax = plt.axes() 
ax.set_xlabel('Time') 
ax.set_ylabel('Work')
plt.plot(a,b)
```

```python
# marker
a = (1,2,3,4,5,6)
b = (3,9,10,12,16,18) 
plt.style.use ('ggplot')
ax = plt.axes()
ax.set(xlabel='Time', ylabel='Work' ) 
plt.plot(a,b,marker='o')
```

```python
# markersize
a = (1,2,3,4,5,6)
b = (3,9,10,12,16,18) 
plt.style.use ('ggplot')
ax = plt.axes()
ax.set(xlabel='Time', ylabel='Work' ) 
plt.plot(a,b,marker='o',markersize=11)
```

```python
# color
a = (1,2,3,4,5,6)
b = (3,9,10,12,16,18) 
plt.style.use ('ggplot')
ax = plt.axes()
ax.set(xlabel='Time', ylabel='Work' ) 
plt.plot(a,b,color='blue',marker='o',markersize=11)
```

```python
# title
a = (1,2,3,4,5,6)
b = (3,9,10,12,16,18) 
plt.style.use ('ggplot')
ax = plt.axes()
ax.set(xlabel='Time', ylabel='Work' ) 
plt.title('time/work consumption') 
plt.plot(a,b,color='blue',marker='o',markersize=11)
```

```python
# exp and sin
import numpy as np
x=np.arange(0,10,0.005)
y = np.exp(-x/2) * np.sin(2*np.pi*x) 
plt.plot (x,y)
```

```python
# Draw Two function
import numpy as np
def p1(x):  
    return x**4 - 4*x**2 + 3*x 
def p2(x):  
    return np.sin(10*x)*10
x = np.linspace(-3,3,1000) 
plt.plot(x,p1(x),x,p2(x))
```

```python
# linestyle
import numpy as np
x=np.arange(0,10,0.005)
plt.plot(x, x + 0, linestyle='solid') 
plt.plot(x, x + 1, linestyle='-')
plt.plot(x, x + 4, linestyle='dashed') 
plt.plot(x, x + 5, linestyle='--')
plt.plot(x, x + 8, linestyle='dashdot') 
plt.plot(x, x + 9, linestyle='-.')
plt.plot(x, x + 12, linestyle='dotted') 
plt.plot(x, x + 13, linestyle=':')
```
